## 4. Create a scatterplot (or plots) for the variables with the strongest correlations and examine the nature of their relationships.

```python
# The code below will create a scatterplot between value_eur variable and release_clause_eur variable
sns.lmplot(x = 'value_eur', y = 'release_clause_eur', data = df_mini_fifa21)
plt.xlabel("value in Euro")
plt.ylabel("release clause in Euro")
```

#### The output above clearly shows that the relationship is very strong, mark by a step correlation lines with most of the data point very close to the line.

```python
# The code below will create a scatterplot between value_eur variable and wage_eur variable
sns.lmplot(x = 'value_eur', y = 'wage_eur', data = df_mini_fifa21)
```

#### The output above also shows a positive relationship. But as indicated by the correlation line steppeness, the relationship is not that strong. Also the point are very dense at a lower value and wage but then start to be more sparse as the value and wage goes up.  There are also couple outliers. A little more investigation would be needed for these two variables.

```python
# The code below will create a scatterplot between value_eur variable and overall_performance variable
sns.lmplot(x = 'value_eur', y = 'overall_performance', data = df_mini_fifa21)
```
