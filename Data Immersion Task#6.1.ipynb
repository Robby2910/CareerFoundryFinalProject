# 04. Data Consistency checks

```python
# the function below will check wether the dataframe contains any mix-type colums
for col in df_fifa21_new.columns.tolist():
  weird = (df_fifa21_new[[col]].applymap(type) != df_fifa21_new[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_fifa21_new[weird]) > 0:
    print (col)
```

```python
# the code below will change the data type for the mix-type columns
df_fifa21_new['club_name'] = df_fifa21_new['club_name'].astype('str')
```

```python
df_fifa21_new['league_name'] = df_fifa21_new['league_name'].astype('str')
```

```python
df_fifa21_new['player_position'] = df_fifa21_new['player_position'].astype('str')
```

```python
df_fifa21_new['joined_team'] = df_fifa21_new['joined_team'].astype('object')
```

```python
# the code below will find missing values in the dataframe
df_fifa21_new.isnull().sum()
```

### The way I decided to handle the missing values is by simply filtering out and removing all the rows with missing value. The rational for that is that the data set is from a web scrapping script ran on the video game FIFA 21. I would then assume that some of the missing data are from players that have been created and that are not real. I would also assume that the data we actually have on hand are for info from real players. Lastly, the removed row would not have a great effect on the data since they will account for less that 5% of the data.
